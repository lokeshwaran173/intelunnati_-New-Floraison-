```python
import matplotlib.pyplot as plt

def plot_locations(locations):
    # Extract longitude and latitude from the list of locations
    longitudes = [loc[0] for loc in locations]
    latitudes = [loc[1] for loc in locations]

    # Plot the locations on the map
    plt.figure(figsize=(10, 6))
    plt.scatter(longitudes, latitudes, color='red', marker='o', label='Locations')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('GIS Software - Plotting Locations')
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    # Example list of locations (longitude, latitude)
    locations = [
        (12.492373, 41.890251),  # Rome, Italy
        (-0.127647, 51.507222),  # London, UK
        (-73.935242, 40.730610), # New York City, USA
        (139.691706, 35.689487)  # Tokyo, Japan
    ]

    plot_locations(locations)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def plot_locations_on_map(locations, clusters=1):
    # Convert the list of locations (longitude, latitude) to a GeoDataFrame
    gdf = gpd.GeoDataFrame(geometry=gpd.points_from_xy([loc[0] for loc in locations], [loc[1] for loc in locations]))

    # Plot the locations on a map using GeoPandas
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    fig, ax = plt.subplots(figsize=(10, 6))
    world.boundary.plot(ax=ax, linewidth=0.8, color='black')
    gdf.plot(ax=ax, color='red', markersize=50, label='Locations')

    # Perform K-Means clustering on the locations (if clusters > 1)
    if clusters > 1:
        kmeans = KMeans(n_clusters=clusters)
        #gdf['cluster'] = kmeans.fit_predict(gdf.geometry)
        #gdf.plot(ax=ax, column='cluster', cmap='Paired', markersize=50, legend=True, label='Clusters')

    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Model GIS Software - Plotting Locations')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    # Example list of locations (longitude, latitude)
    locations = [
        (12.492373, 41.890251),  # Rome, Italy
        (-0.127647, 51.507222),  # London, UK
        (-73.935242, 40.730610), # New York City, USA
        (139.691706, 35.689487)  # Tokyo, Japan
    ]

    # Set the number of clusters for K-Means (1 to plot just the locations without clustering)
    num_clusters = 2

    plot_locations_on_map(locations, num_clusters)
```
